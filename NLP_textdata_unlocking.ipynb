```python
#import urllib.request as urllib2
import requests
import urllib.request as urllib2
from bs4 import BeautifulSoup
```

```python
# fetch the html file
#url = "https://www.coursera.org/learn/language-processing"

#url = "https://www.oreilly.com/library/view/mining-the-social/9781449368180/ch05.html"

url = 'https://en.wikipedia.org/wiki/Natural_language_processing'
response = urllib2.urlopen(url)
```

```python
html_doc =response.read()
```

```python
html_doc
```

```python
# parse the html file
soup =BeautifulSoup(html_doc,"html.parser")
```

```python
soup
```

```python
strhtm=soup.prettify()
```

```python
# extracting all instances of a particular tag

for x in soup.find_all('a'):print(x.string)
```

```python
# extracting all text of a particular tag
for x in soup.find_all('p'):print(x.text)
```
